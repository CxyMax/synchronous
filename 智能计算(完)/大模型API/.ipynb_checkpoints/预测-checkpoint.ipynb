```python
import pandas as pd
import os
import torch
from torch.utils.data import Dataset, DataLoader
from torch import nn
from d2l import torch as d2l
```

### Load Data

```python
col_name = []
unnormal = []
# 绝大部分为34个数据
for i in range(34):
    col_name.append('col'+str(i))
df_price = pd.DataFrame(columns=col_name)
df_price
```

```python
for i,file_name in enumerate(folder_names):
    xls_file = directory_path + '\\' + file_name
    xls_name = os.listdir(xls_file)
    if len(xls_name) == 0:
        continue
    for xls in xls_name:
        xls_path = xls_file + '\\'+ xls
        # 打印路径
        print(xls_path)
        df = pd.read_excel(xls_path)
        price = df.iloc[:,2]
        if len(price) != 34:
            unnormal.append(xls_path)
            continue
        df_price.loc[len(df_price.index)] = price.tolist()
```

### 数据预处理

```python
y = (df_price.iloc[:,-1] > df_price.iloc[:,-2])
y.describe()
```

```python
X_train = df_price.iloc[:800,:-1]
X_test = y[:800].astype(int)
y_train = df_price.iloc[800:,:-1]
y_test = y[800:].astype(int)
```

```python
# 训练集17例正例 验证机10例正例
print((X_test == True).sum(),(y_test == True).sum())
```

```python
class PriceDataset(Dataset):
    def __init__(self, features, labels):
        self.features = torch.tensor(features.values, dtype=torch.float32)
        self.labels = torch.tensor(labels.values, dtype=torch.float32)

    def __len__(self):
        return len(self.features)
        
    def __getitem__(self, idx):
        return self.features[idx], self.labels[idx]
```

```python
# Initialize the dataset
dataset = PriceDataset(X_train, X_test)
dataset_eval = PriceDataset(X_train, X_test)

# Create a DataLoader
batch_size = 32  # You can adjust this value
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)
dataloader_eval = DataLoader(dataset_eval, batch_size=batch_size, shuffle=True)

# Example: Iterating through the DataLoader
for features, labels in dataloader:
    print("Features:", features.shape)
    print("Labels:", labels.shape)
    break  # Only display the first batch
```

### 模型

```python
net = nn.Sequential(nn.Linear(33,64),
                   nn.ReLU(),
                   nn.Linear(64,16),
                   nn.ReLU(),
                   nn.Linear(16,2),
                   nn.Sigmoid())
loss = nn.CrossEntropyLoss(reduction='none')
trainer = torch.optim.SGD(net.parameters(), lr=0.0001)
```

### 训练

```python
epochs = 30
num_epochs = 30

animator = d2l.Animator(xlabel='epoch', xlim=[1, num_epochs], ylim=[0.0, 1],
                        legend=['train loss', 'train acc', 'test acc'])
for epoch in range(epochs):
    
    
    # loss_sumption, accurate_numbers, batch_size for one batch
    train_matrix = d2l.Accumulator(3)
    # 1. 一个epoch的训练
    for X,y in dataloader:
        y = y.long()
        # (1). 训练和参数更新
        if isinstance(net,nn.Module):
            net.train()
        trainer.zero_grad()
        # l = loss(net(X), y)
        y_hat = net(X)
        l = loss(y_hat,y)
        l.mean().backward()
        trainer.step()
        # (2). 一个epoch的loss/accuracy累加
        if isinstance(net,nn.Module):
            net.eval()
        with torch.no_grad():
            train_matrix.add(float(l.sum()), d2l.accuracy(y_hat,y), y.numel())
            
    # 2. 在测试集上的准确率           
    if isinstance(net,nn.Module):
        net.eval()
    with torch.no_grad():
        test_accuracy = d2l.evaluate_accuracy_gpu(net,dataloader_eval)
     # 3. 可视化
        train_metrics = (train_matrix[0]/train_matrix[2], train_matrix[1]/train_matrix[2])
        animator.add(epoch + 1, train_metrics + (test_accuracy,))
```
